1.Positive or Negative
Write a program to check whether a given number is positive, negative, or zero.

```python
num = int(input("Enter a number: "))
if(num>0):
    print(f"Entered number {num} is positive")
elif(num<0):
    print(f"Entered number {num} is negative")
else:
    print(f"Entered number {num} is zero")
```

6. Leap Year Check
Write a program to check whether a given year is a leap year or not. (Hint: A leap year is
divisible by 4, but not by 100 unless also divisible by 400).

```python
year = int(input("Enter a year: "))
if(year%4==0 and (year%100!=0 or year%400==0)):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

9. Character Case Check
Write a program to input a character and check whether it is an uppercase letter,
lowercase letter, or not a letter.

```python
# 9. Character Case Check
# Write a program to input a character and check whether it is an uppercase letter,
# lowercase letter, or not a letter.
ch = input("Enter a character: ")
if(ch.isupper()):
    print(f"Entered character {ch} is an uppercase letter")
elif(ch.islower()):
    print(f"Entered character {ch} is a lowercase letter")
else:
    print(f"Entered character {ch} is not a letter")
```

10. Discount Calculator
Write a program that calculates the discount:
- If purchase amount is greater than or equal to 1000, apply a 10% discount.
- Otherwise, no discount.
Finally, print the final bill amount.

```python
# 10. Discount Calculator
# Write a program that calculates the discount:
# - If purchase amount is greater than or equal to 1000, apply a 10% discount.
# - Otherwise, no discount.
# Finally, print the final bill amount.
amount = float(input("Enter the purchase amount: "))
if(amount>=1000):
    discount = amount * 0.1
    final_amount = amount - discount
    print(f"Purchase amount: {amount}, Discount: {discount}, Final bill amount: {final_amount}")
else:
    print(f"Purchase amount: {amount}, No discount applied, Final bill amount: {amount}")
```
