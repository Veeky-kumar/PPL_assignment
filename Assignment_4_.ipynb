## Q1 Smallest Multiple

```
def find_smallest_multiple(n: int) -> int:
    if n == 1:
        return 1
    x = 1
    while True:
        ok = True
        for i in range(1, n+1):
            if x % i != 0:
                ok = False
                break
        if ok:
            return x
        x += 1

find_smallest_multiple(5)
```

## Q2 Armstrong

```
def cubesum(num: int) -> int:
    return sum(int(d)**3 for d in str(num))

def isArmstrong(num: int) -> bool:
    return cubesum(num) == num

def PrintArmstrong(limit=1000):
    return [i for i in range(limit) if isArmstrong(i)]

PrintArmstrong(500)
```

## Q6 Count chars

```
def count_types(sentence):
    vowels = 'aeiouAEIOU'
    v = c = d = s = 0
    for ch in sentence:
        if ch.isdigit():
            d += 1
        elif ch.isalpha():
            if ch in vowels:
                v += 1
            else:
                c += 1
        else:
            s += 1
    return v, c, d, s

count_types("Hello!! 123")
```
