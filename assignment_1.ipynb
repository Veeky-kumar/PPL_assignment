Q 1) Write a Python program to get a list, sorted in increasing order by the last element in each tuple from a given list of non-empty tuples. 
Sample List: [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)] 

```python
list = [(2,1), (1,2), (2,3), (4,4), (2,5)]
sorted_by_last = sorted(list, key=lambda x: x[1])
print(sorted_by_last)
```

Q2) Write a Python program to get a string from a given string where all occurrences of its first char have been changed to '$', except the first char itself. Sample String: 'restart' 

```python
def replace_char(s):
    first = s[0]
    rest = s[1:]
    rest = rest.replace(first, '$')
    return first + rest

str = "restart"
result = replace_char(str)
print(result)
```

Q3) Write a Python program to find the first appearance of the substring 'not' and 'poor' from a given string, if 'bad' follows the 'poor', replace the whole 'not' ... 'poor' substring with 'good'. Return the resulting string. 

```python
def not_poor_replace(s):
    not_pos = s.find("not")
    poor_pos = s.find("poor")

    print("Index of 'not':", not_pos)
    print("Index of 'poor':", poor_pos)

    if not_pos != -1 and poor_pos != -1 and poor_pos > not_pos:
        s = s[:not_pos] + "good" + s[poor_pos+4:]
    return s


print(not_poor_replace("The movie is not that poor!"))
```

Q4) Write a python program to sort a dictionary by value. 

```python
data = {'apple': 3, 'banana': 1, 'cherry': 2}
sorted_data = sorted(data.items(), key=lambda x: x[1])
print(sorted_data)
```
