## Q1 Classes and Enrollment

```
class Department:
    def __init__(self, name):
        self.name = name
        self.courses = []

    def add_course(self, course):
        self.courses.append(course)

class Course:
    def __init__(self, title, code):
        self.title = title
        self.code = code
        self.students = []

    def enroll(self, student):
        self.students.append(student)

class Student:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll

dept = Department('Computer Science')
course = Course('Python Programming', 'CS101')
dept.add_course(course)
s = Student('Alice', 1)
course.enroll(s)

course.students[0].name
```

## Q2 Inheritance Bill

```
class Bill:
    def __init__(self, amount):
        self.amount = amount

class CashPayment(Bill):
    def pay(self):
        return f"Paid {self.amount} by cash"

class ChequePayment(Bill):
    def __init__(self, amount, cheque_no):
        super().__init__(amount)
        self.cheque_no = cheque_no

    def pay(self):
        return f"Paid {self.amount} by cheque {self.cheque_no}"

CashPayment(500).pay(), ChequePayment(1000, 'CHQ123').pay()
```

## Q3 Overload -=

```
class Distance:
    def __init__(self, meters):
        self.meters = meters

    def __isub__(self, other):
        self.meters -= other.meters
        return self

d1 = Distance(100)
d2 = Distance(40)
d1 -= d2
d1.meters
```

## Q4 Distance Conversion

```
class MeterDist:
    def __init__(self, m):
        self.m = m

    def to_km(self):
        return self.m / 1000

class KmDist:
    def __init__(self, km):
        self.km = km

    def to_m(self):
        return self.km * 1000

MeterDist(500).to_km(), KmDist(2).to_m()
```
