1. Invert a dictionary such that:
- Values become keys
- Multiple original keys mapping to the same value are stored as a sorted list

```python
d = {'a': 1, 'b': 2, 'c': 1, 'd': 2, 'e': 3}
result = {}
for key,value in d.items():
    if value not in result:
        result[value]=[]
    result[value].append(key)

for value in result:
    result[value].sort()

print(result)
```

2.Write a function to get a string made of 4 copies of the last two characters of a specifiedstring (length must be atleast 2).

```python
def last_two_repeat(s):
    if len(s)<2:
        raise ValueError("String len must be greater then 2")
    last_two = s[-2:]
    return (4*last_two)

print(last_two_repeat("python"))
print(last_two_repeat("en"))
# print(last_two_repeat("e"))
```

3.Two vectors are stored as dictionaries of index-value pairs.Write a program to compute their dot product efficiently.

```python
v1 = {0: 1, 2: 3, 4: 5}
v2 = {1: 4, 2: 2, 4: 1}

def dot_product(v1, v2):
    result = 0

    # iterate over smaller dictionary for efficiency
    if len(v1) > len(v2):
        v1, v2 = v2, v1

    for index in v1:
        if index in v2:
            result += v1[index] * v2[index]

    return result

print(dot_product(v1, v2))
```
